```python
import matplotlib.pyplot as plt
import math

x = []
y = []

r = 4
center = [2, 2]

theta_start = 0.0
theta_end = 2 * math.pi
step_size = 0.125

steps = math.ceil((theta_end - theta_start) / step_size)

theta = theta_start
for i in range(steps):
    x.append(r * math.cos(theta) + center[0])
    y.append(r * math.sin(theta) + center[1])
    theta += step_size

# Plot AFTER loop
plt.figure()
plt.plot(center[0], center[1], '.')   # center point
plt.plot(x, y, '-')                  # circle
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")

ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

plt.show()
```

```python
import matplotlib.pyplot as plt
import math

# Parameters
p = 1   # x-coordinate of vertex
q = 2   # y-coordinate of vertex
a = 1   # parabola parameter

# Generate y-values around the vertex
y_values = []
x_values = []

y_start = q - 10
y_end = q + 10
step = 0.05  # smaller step = smoother curve

y = y_start
while y <= y_end:
    y_values.append(y)
    x = ((y - q)**2) / (4 * a) + p
    x_values.append(x)
    y += step

# Plot
plt.plot(x_values, y_values, color='blue', label='Parabola (y-q)^2 = 4a(x-p)')
plt.plot(p, q, 'ro', label='Vertex (p,q)')  # mark vertex
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Parabola with vertex at (p,q)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

# --- Q3a: Sine function ---
x_sin = []
y_sin = []
start = -4 * math.pi
end = 4 * math.pi
step = 0.1
i = start
while i <= end:
    x_sin.append(i)
    y_sin.append(math.sin(i))
    i += step

# --- Q3b: Circle ---
x_circle = []
y_circle = []
r = 4
center_circle = [2, 2]
theta_start = 0.0
theta_end = 2 * math.pi
step_size = 0.125
theta = theta_start
steps = math.ceil((theta_end - theta_start) / step_size)
for _ in range(steps):
    x_circle.append(r * math.cos(theta) + center_circle[0])
    y_circle.append(r * math.sin(theta) + center_circle[1])
    theta += step_size

# --- Q3c: Ellipse ---
x_ellipse = []
y_ellipse = []
a = 3
b = 5
center_ellipse = [2, 2]
theta = theta_start
steps = math.ceil((theta_end - theta_start) / step_size)
for _ in range(steps):
    x_ellipse.append(a * math.cos(theta) + center_ellipse[0])
    y_ellipse.append(b * math.sin(theta) + center_ellipse[1])
    theta += step_size

# --- Create subplots ---
fig, axs = plt.subplots(1, 3, figsize=(18, 5))  # 1 row, 3 columns

# Sine function
axs[0].plot(x_sin, y_sin, color='red')
axs[0].set_title("y = sin(x)")
axs[0].set_xlabel("X-axis")
axs[0].set_ylabel("Y-axis")
axs[0].grid(True)

# Circle
axs[1].plot(center_circle[0], center_circle[1], 'ro')  # center
axs[1].plot(x_circle, y_circle, '-')
axs[1].set_title("Circle with radius 4")
axs[1].set_xlabel("X-axis")
axs[1].set_ylabel("Y-axis")
axs[1].grid(True)
axs[1].set_aspect('equal', adjustable='box')

# Ellipse
axs[2].plot(center_ellipse[0], center_ellipse[1], 'ro')  # center
axs[2].plot(x_ellipse, y_ellipse, '-')
axs[2].set_title("Ellipse a=3, b=5")
axs[2].set_xlabel("X-axis")
axs[2].set_ylabel("Y-axis")
axs[2].grid(True)
axs[2].set_aspect('equal', adjustable='box')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]
enfield = [50,40,70,80,20]
honda   = [80,20,20,50,60]
yahama  = [70,20,60,40,60]
ktm     = [40,20,30,70,60]

fig, axs = plt.subplots(1, 4, figsize=(20,5))  # 1 row, 4 columns

# Enfield
axs[0].plot(days, enfield, marker='o', color='red')
axs[0].set_title("Enfield")
axs[0].set_xlabel("Days")
axs[0].set_ylabel("Distance Covered")
axs[0].grid(True)

# Honda
axs[1].plot(days, honda, marker='s', color='blue')
axs[1].set_title("Honda")
axs[1].set_xlabel("Days")
axs[1].set_ylabel("Distance Covered")
axs[1].grid(True)

# Yahama
axs[2].plot(days, yahama, marker='^', color='green')
axs[2].set_title("Yahama")
axs[2].set_xlabel("Days")
axs[2].set_ylabel("Distance Covered")
axs[2].grid(True)

# KTM
axs[3].plot(days, ktm, marker='d', color='orange')
axs[3].set_title("KTM")
axs[3].set_xlabel("Days")
axs[3].set_ylabel("Distance Covered")
axs[3].grid(True)

# Overall figure title
fig.suptitle("Bike distances over 5 days", fontsize=16)

plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

# Generate x-values from -3π to 3π
x_values = []
y_values = []

start = -3 * math.pi
end = 3 * math.pi
step = 0.1  # smaller step for smoother curve

x = start
while x <= end:
    x_values.append(x)
    y_values.append(math.cos(x))
    x += step

# Plot the cosine function
plt.plot(x_values, y_values, color='blue', label='y = cos(x)')

# Labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Plot of y = cos(x), -3π ≤ x ≤ 3π")
plt.grid(True)
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

# Circle parameters
a = 2
b = 2
r = 4

# Initialize lists for coordinates
x_values = []
y_values = []

# Theta from 0 to 2π
theta = 0
theta_end = 2 * math.pi
step = 0.1  # smaller step for smoother circle

# Generate circle points using while loop
while theta <= theta_end:
    x_values.append(a + r * math.cos(theta))
    y_values.append(b + r * math.sin(theta))
    theta += step

# Plot the circle
plt.plot(x_values, y_values, color='blue', label='Circle')
plt.plot(a, b, 'ro', label='Center')  # mark the center

# Labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Circle with center (2,2) and radius 4")
plt.grid(True)
plt.axis('equal')  # make sure circle is not distorted
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import math

a_circle = 2
b_circle = 2
r_circle = 4

x_circle = []
y_circle = []

theta = 0
theta_end = 2 * math.pi
step = 0.01

while theta <= theta_end:
    x_circle.append(a_circle + r_circle * math.cos(theta))
    y_circle.append(b_circle + r_circle * math.sin(theta))
    theta += step

a_ellipse = 3
b_ellipse = 5
center_ellipse = [2, 2]

x_ellipse = []
y_ellipse = []

theta = 0
theta_end = 2 * math.pi

while theta <= theta_end:
    x_ellipse.append(center_ellipse[0] + a_ellipse * math.cos(theta))
    y_ellipse.append(center_ellipse[1] + b_ellipse * math.sin(theta))
    theta += step

plt.plot(x_circle, y_circle, color='blue', label='Circle')
plt.plot(x_ellipse, y_ellipse, color='red', label='Ellipse')
plt.plot(a_circle, b_circle, 'ko', label='Center')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Circle and Ellipse on Same Axes")
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```
