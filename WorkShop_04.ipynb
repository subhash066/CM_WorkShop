```python
import matplotlib.pyplot as plt
import random

# Generate 10 random x and y values between 0 and 50
x = [random.randint(0, 50) for _ in range(10)]
y = [random.randint(0, 50) for _ in range(10)]

# Define colors for each point
colors = ["red", "green", "blue", "yellow", "pink",
          "orange", "purple", "brown", "cyan", "magenta"]

# Create scatter plot
plt.scatter(x, y, c=colors)

# Label axes
plt.xlabel("Time taken in seconds")
plt.ylabel("Distance reached")

# Add title
plt.title("Buzzer Game Results")

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example population data (replace with actual values if given)
# Format: {year: [population_of_animals]}
data = {
    2017: [50, 30, 20],
    2018: [60, 25, 15],
    2019: [55, 35, 10],
    2020: [70, 20, 10],
    2021: [65, 25, 10],
    2022: [80, 15, 5]
}

# Labels for animals
animals = ["Tiger", "Elephant", "Deer"]

# Create subplots (2 rows, 3 columns for 6 years)
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

# Flatten axes for easy iteration
axes = axes.flatten()

# Loop through each year and plot pie chart
for i, (year, populations) in enumerate(data.items()):
    axes[i].pie(populations, labels=animals, autopct='%1.1f%%')
    axes[i].set_title(f"Population in {year}")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
car_colors = ["Red", "Silver", "Blue", "White", "Green"]
number_seen = [123, 230, 78, 193, 12]

# (a) Basic bar chart
plt.bar(car_colors, number_seen)
plt.title("Car Colors Seen")
plt.xlabel("Car Colors")
plt.ylabel("Number Seen")
plt.show()

# (b) Bar chart with width = 0.3
plt.bar(car_colors, number_seen, width=0.3)
plt.title("Bar Chart with Width = 0.3")
plt.show()
# Effect: Bars become thinner (narrower).

# (c) Horizontal bar chart
plt.barh(car_colors, number_seen)
plt.title("Horizontal Bar Chart")
plt.show()
# Effect: Bars are displayed horizontally instead of vertically.

# (d) Bar chart with matching colors
plt.bar(car_colors, number_seen, color=["red", "silver", "blue", "white", "green"])
plt.title("Car Colors with Matching Bar Colors")
plt.show()

# (e) Pie chart of car data
plt.pie(number_seen, labels=car_colors, autopct='%1.1f%%')
plt.title("Car Colors Pie Chart")
plt.show()

# (f) Testing additional pie chart arguments/settings

# A: Explode one slice
myexplode = [0.3, 0, 0, 0, 0]
plt.pie(number_seen, labels=car_colors, autopct='%1.1f%%', explode=myexplode)
plt.title("Pie Chart with Exploded Slice")
plt.show()

# B: Explode with shadow
plt.pie(number_seen, labels=car_colors, autopct='%1.1f%%', explode=myexplode, shadow=True)
plt.title("Pie Chart with Exploded Slice + Shadow")
plt.show()

# C: Add legend
plt.pie(number_seen, labels=car_colors, autopct='%1.1f%%')
plt.legend(title="Car Colours:")
plt.title("Pie Chart with Legend")
plt.show()

# D: Custom colors
mycolours = ["Magenta", "Green", "Cyan", "Red", "Yellow"]
plt.pie(number_seen, labels=car_colors, autopct='%1.1f%%', colors=mycolours)
plt.title("Pie Chart with Custom Colours")
plt.show()
```
