```python
def unique_elements(input_list):
    unique_list = []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list


sample_list = [1, 2, 3, 3, 3, 3, 4, 5]
print("Unique List:", unique_elements(sample_list))
```

```python
def is_prime(n):
    # Prime numbers are greater than 1
    if n <= 1:
        return False

    # Check divisibility from 2 up to sqrt(n)
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False

    return True

# Example usage
print(is_prime(2))   # True
print(is_prime(4))   # False
print(is_prime(17))  # True
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def twin_primes(limit):
    primes = [p for p in range(2, limit) if is_prime(p)]
    twins = []
    for i in range(len(primes) - 1):
        if primes[i+1] - primes[i] == 2:
            twins.append((primes[i], primes[i+1]))
    return twins

# Example usage
print(twin_primes(1000))
```

```python
# Function to calculate sum of divisors (excluding the number itself)
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:   # check if i divides n
            total += i
    return total

# Function to find the next perfect number after 6
def next_perfect_number(start=7):
    n = start
    while True:
        if sumOfDivisors(n) == n:   # check if n is perfect
            return n
        n += 1

# Print the next perfect number after 6
print(next_perfect_number())
```

```python
def decimal_to_binary(n):
    binary = ""
    while n > 0:
        remainder = n % 2      # remainder (modulas)
        binary = str(remainder) + binary
        n = n // 2
    return binary if binary != "" else "0"

# Taking input manually
num = int(input("Enter a decimal number: "))
print("Binary of", num, "is:", decimal_to_binary(num))
```

```python
def prodDigits(n):
    product = 1
    while n > 0:
        digit = n % 10       # extract last digit
        product *= digit     # multiply with product
        n = n // 10          # remove last digit
    return product

num =int(input("Enter a digit: "))
print("Product of digits of", num, "is:", prodDigits(num))
```
